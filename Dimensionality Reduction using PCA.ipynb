```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
import pandas as pd
features_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
target_df = pd.DataFrame(cancer.target, columns=["breast_cancer"])
```

```python
from sklearn.preprocessing import StandardScaler
# Standardizing the features
scaler = StandardScaler()
scaler.fit(features_df)
features_df = scaler.transform(features_df)
```

```python
features_df = pd.DataFrame(features_df, columns=cancer.feature_names)
.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(features_df)
```

```python
import numpy as np

tot_var = np.array(pca.explained_variance_ratio_)
print(tot_var)
```

```python
tot_var.cumsum()
```

```python
pca = PCA(n_components = 5) 
pca.fit(features_df)
features_df_pca = pca.transform(features_df)
```

```python
features_df_pca = pd.DataFrame(features_df_pca, columns=["Comp1", "Comp2", "Comp3", "Comp4", "Comp5"])
```

```python
df = pd.concat([features_df_pca, target_df], axis=1)
df.head()
```

```python
import seaborn as sbn

sbn.pairplot(data=df, hue="breast_cancer")
```
