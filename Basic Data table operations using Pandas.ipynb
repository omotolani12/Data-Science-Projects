```python
from sklearn.datasets import load_iris
data = load_iris()

import pandas as pd

iris = pd.DataFrame(data = data.data, columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"])
iris['species'] = pd.Categorical.from_codes(data.target, data.target_names)
```

```python
iris.head()
```

```python
query = iris.loc[(iris.petal_length>4) & (iris.species=="versicolor"),
                  ["sepal_length", "sepal_width"]]

query.head()
```
